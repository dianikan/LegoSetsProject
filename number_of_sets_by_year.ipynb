```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data' is your DataFrame and 'year' is the column
year_counts = data['year'].value_counts().sort_index()

# Create a line chart
plt.figure(figsize=(12, 6))  # Increase figure size
year_counts.plot(kind='line', color='steelblue')  # Change line color

plt.title('Distribution of Years')
plt.xlabel('Year')
plt.ylabel('Count')

# Add annotations for min and max values
min_year = year_counts.idxmin()
min_value = year_counts.min()
max_year = year_counts.idxmax()
max_value = year_counts.max()

plt.text(min_year, min_value + 10, f'{min_value}', ha='center', va='bottom')  # Adjust text position
plt.plot(min_year, min_value, 'bo', markersize=8)  # Increase marker size

plt.text(max_year, max_value - 20, f'{max_value}', ha='center', va='top')  # Adjust text position
plt.plot(max_year, max_value, 'bo', markersize=8)  # Increase marker size

plt.xticks(range(year_counts.index.min(), year_counts.index.max() + 5, 5)) #Add more ticks to the x axis.
plt.grid(True, linestyle='--', alpha=0.7)  # Add gridlines

plt.tight_layout() #prevent labels from being cutoff.
plt.show()
```
